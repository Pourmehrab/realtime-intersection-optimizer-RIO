```python
import os
import pandas as pd
import glob
```

```python
inter_name = 'reserv'

filepath = os.path.join('data/' + inter_name + '.csv')
if os.path.exists(filepath):
    df = pd.read_csv(filepath)
else:
    raise Exception(filepath + ' was not found.')
```

```python
# DECOMPOSE SCENARIOS TO DIFFERENT CSV FILES
for sc in df.sc.unique():
    filepath = os.path.join('data/' + inter_name + '_' + str(sc) + '.csv')
    df[df['sc'] == sc].to_csv(filepath, index=False)
```

```python
# RECOMPOSE THE DATAFRAMAE
path = os.path.join('data/')
allFiles = glob.glob(path + "/*.csv")
frame = pd.DataFrame()
list_ = []
for file_ in allFiles:
    df = pd.read_csv(file_, index_col=None, header=0)
    list_.append(df)
frame = pd.concat(list_)
frame.to_csv(path + '_joined.csv', index=False)
```
